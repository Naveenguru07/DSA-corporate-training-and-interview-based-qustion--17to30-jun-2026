```python
class Solution:
    def trap(self, height):

        start = 0
        end = len(height) - 1

        leftMax = 0
        rightMax = 0
        totalWater = 0

        while start < end:

            leftMax = max(leftMax, height[start])
            rightMax = max(rightMax, height[end])

            if leftMax < rightMax:
                totalWater += leftMax - height[start]
                start += 1
            else:
                totalWater += rightMax - height[end]
                end -= 1

        return totalWater

# Object create
obj = Solution()

# Function call
height = [0,1,0,2,1,0,1,3,2,1,2,1]
print(obj.trap(height))
```

```python
#First missing possitive
class Solution:
    def firstMissingPositive(self, nums):
        nums = [n for n in nums if n > 0]
        nums.sort()

        target = 1
        for n in nums:
            if n == target:
                target += 1
            elif n > target:
                return target

        return target

# User input
nums = list(map(int, input("Enter array elements: ").split()))

obj = Solution()
print("Output:", obj.firstMissingPositive(nums))
```

```python
#Sliding Window Maximum
from collections import deque

class Solution:
    def maxSlidingWindow(self, nums, k):
        res = []
        q = deque()

        for idx, num in enumerate(nums):
            while q and q[-1] < num:
                q.pop()

            q.append(num)

            if idx >= k and nums[idx-k] == q[0]:
                q.popleft()

            if idx >= k-1:
                res.append(q[0])

        return res

# User Input
nums = list(map(int, input("Enter array elements: ").split()))
k = int(input("Enter window size (k): "))

obj = Solution()
print("Output:", obj.maxSlidingWindow(nums, k))
```

```python
#Median of Two Sorted Arrays
class Solution:
    def findMedianSortedArrays(self, nums1, nums2):

        merged = sorted(nums1 + nums2)
        length = len(merged)


        if length % 2 == 0:
            return (merged[length // 2 - 1] + merged[length // 2]) / 2


        else:
            return merged[length // 2]

# User Input
nums1 = list(map(int, input("Enter first sorted array: ").split()))
nums2 = list(map(int, input("Enter second sorted array: ").split()))

obj = Solution()
print("Median =", obj.findMedianSortedArrays(nums1, nums2))
```

```python
#Maximum Sum of 3 Non-Overlapping Subarrays
class Solution:
    def maxSumOfThreeSubarrays(self, nums, k):
        maxSum = 0
        ans = []

        n = len(nums)

        for i in range(n - 3 * k + 1):
            sum1 = sum(nums[i:i+k])

            for j in range(i + k, n - 2 * k + 1):
                sum2 = sum(nums[j:j+k])

                for l in range(j + k, n - k + 1):
                    sum3 = sum(nums[l:l+k])

                    total = sum1 + sum2 + sum3

                    if total > maxSum:
                        maxSum = total
                        ans = [i, j, l]

        return ans


# User Input
nums = list(map(int, input("Enter array elements: ").split()))
k = int(input("Enter k: "))

obj = Solution()
print("Output:", obj.maxSumOfThreeSubarrays(nums, k))
```

```python
#Minimum Window Substring
class Solution:
    def minWindow(self, s, t):
        ans = ""

        for i in range(len(s)):
            for j in range(i + 1, len(s) + 1):
                sub = s[i:j]

                flag = True
                for ch in t:
                    if sub.count(ch) < t.count(ch):
                        flag = False
                        break

                if flag:
                    if ans == "" or len(sub) < len(ans):
                        ans = sub

        return ans


# User Input
s = input("Enter main string: ")
t = input("Enter target string: ")

obj = Solution()
print("Output:", obj.minWindow(s, t))
```

```python
# Substring with Concatenation of All Words
class Solution:
    def findSubstring(self, s, words):
        result = []

        word_len = len(words[0])
        total_len = word_len * len(words)

        for i in range(len(s) - total_len + 1):
            temp = []

            for j in range(i, i + total_len, word_len):
                temp.append(s[j:j + word_len])

            if sorted(temp) == sorted(words):
                result.append(i)

        return result


# User Input
s = input("Enter the string: ")
words = input("Enter words (space separated): ").split()

obj = Solution()
print("Output:", obj.findSubstring(s, words))
```

```python
#Regular Expression Matching
class Solution:
    def isMatch(self, s, p):

        if not p:
            return s == ""


        first = len(s) > 0 and (s[0] == p[0] or p[0] == ".")


        if len(p) >= 2 and p[1] == "*":
            return self.isMatch(s, p[2:]) or (first and self.isMatch(s[1:], p))

        return first and self.isMatch(s[1:], p[1:])


# User Input
s = input("Enter string: ")
p = input("Enter pattern: ")

obj = Solution()
print("Output:", obj.isMatch(s, p))
```

```python
# Palindrome Partitioning II
class Solution:
    def isPalindrome(self, s):
        return s == s[::-1]

    def minCut(self, s):
        if self.isPalindrome(s):
            return 0

        n = len(s)
        minCuts = n - 1

        for i in range(1, n):
            left = s[:i]
            right = s[i:]

            if self.isPalindrome(left) and self.isPalindrome(right):
                minCuts = 1

        return minCuts


# User Input
s = input("Enter a string: ")

obj = Solution()
print("Output:", obj.minCut(s))
```

```python
#Distinct Subsequences
class Solution:
    def numDistinct(self, s, t):


        if t == "":
            return 1

        if s == "":
            return 0

        if s[0] == t[0]:
            return self.numDistinct(s[1:], t[1:]) + self.numDistinct(s[1:], t)


        return self.numDistinct(s[1:], t)


# User Input
s = input("Enter source string: ")
t = input("Enter target string: ")

obj = Solution()
print("Output:", obj.numDistinct(s, t))
```

```python
# Largest Rectangle in Histogram
class Solution:
    def largestRectangleArea(self, heights):
        maxArea = 0
        n = len(heights)

        for i in range(n):
            minHeight = heights[i]

            for j in range(i, n):
                minHeight = min(minHeight, heights[j])
                area = minHeight * (j - i + 1)

                if area > maxArea:
                    maxArea = area

        return maxArea


# User Input
heights = list(map(int, input("Enter heights: ").split()))

obj = Solution()
print("Output:", obj.largestRectangleArea(heights))
```

```python
# Shortest Subarray with Sum at Least K
class Solution:
    def shortestSubarray(self, nums, k):
        n = len(nums)
        ans = n + 1

        for i in range(n):
            total = 0
            for j in range(i, n):
                total += nums[j]

                if total >= k:
                    ans = min(ans, j - i + 1)
                    break

        if ans == n + 1:
            return -1
        return ans


# User Input
nums = list(map(int, input("Enter array elements: ").split()))
k = int(input("Enter value of k: "))

obj = Solution()
print("Output:", obj.shortestSubarray(nums, k))
```

```python
# Constrained Subsequence Sum
class Solution:
    def constrainedSubsetSum(self, nums, k):
        maxSum = nums[0]

        for i in range(len(nums)):
            total = 0

            for j in range(i, min(i + k, len(nums))):
                total += nums[j]

                if total > maxSum:
                    maxSum = total

        return maxSum


# User Input
nums = list(map(int, input("Enter array elements: ").split()))
k = int(input("Enter value of k: "))

obj = Solution()
print("Output:", obj.constrainedSubsetSum(nums, k))
```

```python
# Jump Game VI
class Solution:
    def maxResult(self, nums, k):
        n = len(nums)
        dp = [0] * n
        dp[0] = nums[0]

        for i in range(1, n):
            best = float('-inf')

            for j in range(max(0, i - k), i):
                best = max(best, dp[j])

            dp[i] = nums[i] + best

        return dp[-1]


# User Input
nums = list(map(int, input("Enter array elements: ").split()))
k = int(input("Enter value of k: "))

obj = Solution()
print("Output:", obj.maxResult(nums, k))
```

```python
# Reverse Nodes in k-Group
class Solution:
    def reverseKGroup(self, nums, k):
        result = []

        for i in range(0, len(nums), k):
            group = nums[i:i+k]

            if len(group) == k:
                result.extend(group[::-1])
            else:
                result.extend(group)

        return result


# User Input
nums = list(map(int, input("Enter array elements: ").split()))
k = int(input("Enter value of k: "))

obj = Solution()
print("Output:", obj.reverseKGroup(nums, k))
```

```python
# Median of Two Sorted Arrays
class Solution:
    def findMedianSortedArrays(self, nums1, nums2):

        merged = nums1 + nums2
        merged.sort()

        n = len(merged)

        if n % 2 == 0:
            median = (merged[n//2 - 1] + merged[n//2]) / 2
        else:
            median = merged[n//2]

        return median


# User Input
nums1 = list(map(int, input("Enter first sorted array: ").split()))
nums2 = list(map(int, input("Enter second sorted array: ").split()))

obj = Solution()
print("Output:", obj.findMedianSortedArrays(nums1, nums2))
```

```python
# Split Array Largest Sum
class Solution:
    def splitArray(self, nums, k):
        n = len(nums)
        ans = sum(nums)

        for i in range(n):
            total = 0
            count = 1

            for j in range(n):
                if total + nums[j] <= ans:
                    total += nums[j]
                else:
                    count += 1
                    total = nums[j]

            if count <= k:
                ans = min(ans, total)

        return ans


# User Input
nums = list(map(int, input("Enter array elements: ").split()))
k = int(input("Enter value of k: "))

obj = Solution()
print("Output:", obj.splitArray(nums, k))
```

```python
# Find Peak Element
class Solution:
    def findPeakElement(self, nums):
        n = len(nums)

        for i in range(n):
            if (i == 0 or nums[i] > nums[i-1]) and (i == n-1 or nums[i] > nums[i+1]):
                return i


# User Input
nums = list(map(int, input("Enter array elements: ").split()))

obj = Solution()
print("Peak Element Index:", obj.findPeakElement(nums))
```

```python
# Merge Intervals
class Solution:
    def merge(self, intervals):
        intervals.sort()
        result = []

        for interval in intervals:
            if not result or result[-1][1] < interval[0]:
                result.append(interval)
            else:
                result[-1][1] = max(result[-1][1], interval[1])

        return result


# User Input
n = int(input("Enter number of intervals: "))

intervals = []
for i in range(n):
    start, end = map(int, input("Enter interval: ").split())
    intervals.append([start, end])

obj = Solution()
print("Output:", obj.merge(intervals))
```

```python
# Russian Doll Envelopes
class Solution:
    def maxEnvelopes(self, envelopes):

        # Width அடிப்படையில் sort
        envelopes.sort()

        count = 1

        for i in range(1, len(envelopes)):
            if (envelopes[i][0] > envelopes[i-1][0] and
                envelopes[i][1] > envelopes[i-1][1]):
                count += 1

        return count


# User Input
n = int(input("Enter number of envelopes: "))

envelopes = []
for i in range(n):
    w, h = map(int, input("Enter width and height: ").split())
    envelopes.append([w, h])

obj = Solution()
print("Output:", obj.maxEnvelopes(envelopes))
```

```python
#Maximum Gap
class Solution:
    def maximumGap(self, nums):

        if len(nums) < 2:
            return 0

        nums.sort()

        maxGap = 0

        for i in range(1, len(nums)):
            gap = nums[i] - nums[i-1]

            if gap > maxGap:
                maxGap = gap

        return maxGap


# User Input
nums = list(map(int, input("Enter array elements: ").split()))

obj = Solution()
print("Output:", obj.maximumGap(nums))
```

```python
# House Robber III
class Solution:
    def rob(self, nums):
        if len(nums) == 0:
            return 0
        if len(nums) == 1:
            return nums[0]

        dp = [0] * len(nums)
        dp[0] = nums[0]
        dp[1] = max(nums[0], nums[1])

        for i in range(2, len(nums)):
            dp[i] = max(dp[i-1], dp[i-2] + nums[i])

        return dp[-1]


# User Input
nums = list(map(int, input("Enter house values: ").split()))

obj = Solution()
print("Maximum Amount:", obj.rob(nums))
```

```python
# Binary Tree Maximum Path Sum
class Solution:
    def maxPathSum(self, nums):
        maxSum = nums[0]
        currentSum = 0

        for num in nums:
            currentSum += num

            if currentSum > maxSum:
                maxSum = currentSum

            if currentSum < 0:
                currentSum = 0

        return maxSum


# User Input
nums = list(map(int, input("Enter node values: ").split()))

obj = Solution()
print("Maximum Path Sum:", obj.maxPathSum(nums))
```

```python
# Lowest Common Ancestor of a Binary Tree III
class Solution:
    def lowestCommonAncestor(self, parent, p, q):

        while p != q:
            if p > q:
                p = parent[p]
            else:
                q = parent[q]

        return p


# User Input
n = int(input("Enter number of nodes: "))

parent = {}
print("Enter node and its parent (-1 for root):")
for i in range(n):
    node, par = map(int, input().split())
    parent[node] = par

p = int(input("Enter first node: "))
q = int(input("Enter second node: "))

obj = Solution()
print("Lowest Common Ancestor:", obj.lowestCommonAncestor(parent, p, q))
```

```python
# Alien Dictionary
class Solution:
    def alienOrder(self, words):

        letters = set()

        for word in words:
            for ch in word:
                letters.add(ch)

        result = sorted(letters)

        return "".join(result)


# User Input
words = input("Enter words (space separated): ").split()

obj = Solution()
print("Alien Dictionary Order:", obj.alienOrder(words))
```

```python
# Shortest Path Visiting All Nodes
class Solution:
    def shortestPathLength(self, graph):
        visited = set()
        queue = [(0, 0)]   # (node, distance)
        visited.add(0)

        while queue:
            node, dist = queue.pop(0)

            for neighbor in graph[node]:
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append((neighbor, dist + 1))

        return dist


# User Input
n = int(input("Enter number of nodes: "))

graph = []
print("Enter adjacent nodes for each node (space separated):")
for i in range(n):
    row = list(map(int, input().split()))
    graph.append(row)

obj = Solution()
print("Shortest Path Length:", obj.shortestPathLength(graph))
```

```python
# Longest Consecutive Sequence
class Solution:
    def longestConsecutive(self, nums):

        if len(nums) == 0:
            return 0

        nums.sort()

        longest = 1
        count = 1

        for i in range(1, len(nums)):
            if nums[i] == nums[i-1]:
                continue
            elif nums[i] == nums[i-1] + 1:
                count += 1
            else:
                longest = max(longest, count)
                count = 1

        longest = max(longest, count)
        return longest


# User Input
nums = list(map(int, input("Enter array elements: ").split()))

obj = Solution()
print("Output:", obj.longestConsecutive(nums))
```

```python
#Subarray Sum Equals K
class Solution:
    def subarraySum(self, nums, k):
        count = 0

        for i in range(len(nums)):
            total = 0

            for j in range(i, len(nums)):
                total += nums[j]

                if total == k:
                    count += 1

        return count


# User Input
nums = list(map(int, input("Enter array elements: ").split()))
k = int(input("Enter value of k: "))

obj = Solution()
print("Output:", obj.subarraySum(nums, k))
```

```python
# Maximum Size Subarray Sum Equals K
class Solution:
    def maxSubArrayLen(self, nums, k):
        maxLen = 0

        for i in range(len(nums)):
            total = 0

            for j in range(i, len(nums)):
                total += nums[j]

                if total == k:
                    maxLen = max(maxLen, j - i + 1)

        return maxLen


# User Input
nums = list(map(int, input("Enter array elements: ").split()))
k = int(input("Enter value of k: "))

obj = Solution()
print("Output:", obj.maxSubArrayLen(nums, k))
```

```python
# Minimum Window Substring
class Solution:
    def minWindow(self, s, t):

        minStr = ""

        for i in range(len(s)):
            for j in range(i + 1, len(s) + 1):
                sub = s[i:j]

                ok = True
                for ch in t:
                    if ch not in sub:
                        ok = False
                        break

                if ok:
                    if minStr == "" or len(sub) < len(minStr):
                        minStr = sub

        return minStr


# User Input
s = input("Enter string: ")
t = input("Enter target string: ")

obj = Solution()
print("Output:", obj.minWindow(s, t))
```

```python
class Solution:
    def groupAnagrams(self, strs):
        result = []

        while strs:
            word = strs[0]
            group = [word]
            strs.remove(word)

            for s in strs[:]:
                if sorted(word) == sorted(s):
                    group.append(s)
                    strs.remove(s)

            result.append(group)

        return result


# User Input
strs = input("Enter words (space separated): ").split()

obj = Solution()
print("Output:", obj.groupAnagrams(strs))
```

```python
# N-Queens
class Solution:
    def solveNQueens(self, n):
        board = []

        for i in range(n):
            row = ["."] * n
            row[i] = "Q"
            board.append("".join(row))

        return board


# User Input
n = int(input("Enter value of n: "))

obj = Solution()
result = obj.solveNQueens(n)

print("Output:")
for row in result:
    print(row)
```

```python
# Sudoku Solver
class Solution:
    def solveSudoku(self, board):
        def solve():
            for row in range(9):
                for col in range(9):
                    if board[row][col] == '.':
                        for num in '123456789':
                            if is_valid(row, col, num):
                                board[row][col] = num
                                if solve():
                                    return True
                                board[row][col] = '.'
                        return False
            return True

        def is_valid(row, col, num):
            for i in range(9):
                if board[row][i] == num or board[i][col] == num:
                    return False
            start_row, start_col = 3 * (row // 3), 3 * (col // 3)
            for i in range(start_row, start_row + 3):
                for j in range(start_col, start_col + 3):
                    if board[i][j] == num:
                        return False
            return True

        solve()
```

```python
# Palindrome Partitioning
class Solution:
    def partition(self, s):
        result = []

        for i in range(len(s)):
            for j in range(i + 1, len(s) + 1):
                sub = s[i:j]
                if sub == sub[::-1]:
                    result.append(sub)

        return result


# User Input
s = input("Enter string: ")

obj = Solution()
print("Output:", obj.partition(s))
```

```python
# Word Search II
class Solution:
    def findWords(self, board, words):
        result = []

        for word in words:
            found = False

            for row in board:
                s = "".join(row)
                if word in s:
                    found = True

            if found:
                result.append(word)

        return result


# User Input
r = int(input("Enter number of rows: "))
c = int(input("Enter number of columns: "))

board = []
print("Enter board:")
for i in range(r):
    board.append(input().split())

words = input("Enter words (space separated): ").split()

obj = Solution()
print("Output:", obj.findWords(board, words))
```
